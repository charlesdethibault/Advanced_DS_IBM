### Seasonality Analysis

* We can see that over time the number of sales items is decreasing Year on Year
* The peak period for sales seems to be in the month of December and January and tend to decrease gradually until the month of December of next Year
* The best sales day is on the Saturday. Weekends are the two best days of the week

Seasonality is an important factor in the optimisation. The creation of days of week data, monthly data, and year data makes therefore sense. 

```python
cats = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday']
# Note: Seaborn allows group by but it is much slower so not efficient
Date = df.groupby('date')[
    'item_cnt_day'].count().reset_index()
Date['Month'] = Date['date'].dt.month
Date['Year'] = Date['date'].dt.year
Date['DOWname'] = Date['date'].dt.weekday_name
```

```python
sns.set(style="white")
f,axes = plt.subplots(1,3,figsize=(20,7),sharex=False)
ax1 = sns.lineplot(x="date",y="item_cnt_day", data=Date,ax=axes[0])
ax1.set_title("Item count per day whole period")
ax1 = sns.pointplot(x="Month", y="item_cnt_day",hue="Year", data=Date,ax=axes[1])
ax1.set_title("item_cnt_day per Month per year")
ax2 = sns.pointplot(x="DOWname", y="item_cnt_day",hue="Year", data=Date,ax=axes[2]);
ax2.set_title("item_cnt_day per Day of Week per year");
```

```python
# create a weekend variable
df['day'] = pd.to_datetime(df['date']).dt.day
df['weekday'] = pd.to_datetime(df['date']).dt.dayofweek  # monday = 0, sunday = 6
df['weekend_indi'] = 0          # Initialize the column with default value of 0
df.loc[df['weekday'].isin([5, 6]), 'weekend_indi'] = 1 
```

## Catboost features preparation

```python
# drop columns not to be used and set data type
dfcat = df
dfcat = dfcat.drop(['item_name','DOWname','YYYYMM','shop_name'], axis = 1)
for col in ['shop_id', 'item_id'
            ,'item_category_id','Month','Year','weekday','weekend_indi','city']:
    dfcat[col] = dfcat[col].astype('category')
dfcat.dtypes
dfcat.head()

categorical = ['shop_id','item_id','item_category_id'
                ,'Month','Year','weekday','weekend_indi','city']
```

```python
X = dfcat.drop('item_cnt_day', axis=1)
y = dfcat.item_cnt_day
def column_index(dfcat, query_cols):
    cols = dfcat.columns.values
    sidx = np.argsort(cols)
    return sidx[np.searchsorted(cols, query_cols, sorter=sidx)]
categorical_features_indices = column_index(X, categorical)
```

```python
categorical
categorical_features_indices
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.25, random_state=42)
model=CatBoostRegressor(iterations=200, depth=4, learning_rate=0.1, loss_function='RMSE')
model.fit(X_train, y_train,cat_features=categorical_features_indices,eval_set=(X_valid, y_valid),plot=True);
```

```python
fea_imp = pd.DataFrame({'imp': model.feature_importances_, 'col': X.columns})
fea_imp = fea_imp.sort_values(['imp', 'col'], ascending=[True, False]).iloc[-30:]
fea_imp.plot(kind='barh', x='col', y='imp', figsize=(10, 7), legend=None)
plt.title('CatBoost - Feature Importance')
plt.ylabel('Features')
plt.xlabel('Importance');
```
